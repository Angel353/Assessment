```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("DataBase.xlsx")
df.info()
```

```python
df.head(10)
```

```python
#Identificar valores nulos por columna
valores_nulos1= df.isnull().sum()
valores_nulos1
```

```python
# Hacemos una copia de un data frame limpio
BaseLimpia = df.copy()
```

```python
# Separamos los datos cuantitativos también
BaseCuanti = BaseLimpia.select_dtypes(np.number)
BaseCuanti
```

```python
# Elegimos solo los datos cualitativos y los separamos de los datos cuantitativos.
BaseCuali = BaseLimpia.select_dtypes(include=['object'])
BaseCuali
```

```python
# Reemplazamos valores nulos cualitativos en el dataframe con "desconocido" puesto que no existe ningun patrón para encontrar los datos exactos

CualiLimpia = BaseCuali.fillna("Desconocido") 
CualiLimpia
```

```python
#Identificar valores nulos por columna
valores_nulos2 = CualiLimpia.isnull().sum()
valores_nulos2
```

```python
# Limpieza de datos cuantitativos

# Sustituir valores nulos con promedio 
BaseCuanti["distanciacasa"]= BaseCuanti["distanciacasa"].fillna(round(BaseCuanti["distanciacasa"].mean(),1))
# Hicimos media con redondeo debido a que las variables están en formato de Escala Likert por lo que no fue necesario aplicar mediana, sí hay parametros.
BaseCuanti["grado"]=BaseCuanti["grado"].fillna(round(BaseCuanti["grado"].mean(),1))
BaseCuanti["satisfaccion"]=BaseCuanti["satisfaccion"].fillna(round(BaseCuanti["satisfaccion"].mean(),1))
BaseCuanti["desemp"]=BaseCuanti["desemp"].fillna(round(BaseCuanti["desemp"].mean(),1))
BaseCuanti["balancevida"]=BaseCuanti["balancevida"].fillna(round(BaseCuanti["balancevida"].mean(),1))

# Para estas variables númericas hicimos mediana, ya que se evita la sensibilidad a valores extremos
BaseCuanti["sueldo"]=BaseCuanti["sueldo"].fillna(round(BaseCuanti["sueldo"].median(),1))
BaseCuanti["capacitacion"]=BaseCuanti["capacitacion"].fillna(round(BaseCuanti["capacitacion"].median(),1))
BaseCuanti["antiguedad"]=BaseCuanti["antiguedad"].fillna(round(BaseCuanti["antiguedad"].median(),1))
BaseCuanti["antiguedadpuesto"]=BaseCuanti["antiguedadpuesto"].fillna(round(BaseCuanti["antiguedadpuesto"].median(),1))
BaseCuanti["promocion"]=BaseCuanti["promocion"].fillna(round(BaseCuanti["promocion"].median(),1))
```

```python
# Verificamos valores atípicos por plot (diagrama de caja o bigote) de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
BaseCuanti.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
# Realizamos una prueba al cortar el Data Frame y ver los outliers

BaseOut = BaseCuanti[['tiempojefe','promocion','antiguedadpuesto','antiguedad','capacitacion']]
BaseOut
```

```python
# Verificamos valores atípicos por plot (diagrama de caja o bigote) de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
BaseOut.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
# Guardamos las medidas que sí tienen outliers en otro data frame

BaseOutlier = BaseCuanti[['capacitacion']]
BaseOutlier

# Las dropeamos del data frame original

BaseNoOut = BaseCuanti.drop(['capacitacion'], axis=1)
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y= BaseOutlier

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= BaseOutlier[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= BaseOutlier[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos = data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "median"
#Realizamos una copia del dataframe
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.median(),1))
data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO sin outliers
valores_nulos_4 = data4_iqr.isnull().sum()
valores_nulos_4
```

```python
# Unimos las columnas cualitativas y las que dado su naturaleza no poseían outliers con el dataframe

CuantiAll = pd.concat([BaseNoOut, data4_iqr], axis=1)
CuantiAll
```

```python
# Unimos las columnas cualitativas y las que dado su naturaleza no poseían outliers con el dataframe

BaseCompleta = pd.concat([CuantiAll, CualiLimpia], axis=1)
BaseCompleta
```

```python
#Identificar valores nulos por columna
valores_nulos_all = BaseCompleta.isnull().sum()
valores_nulos_all
```
